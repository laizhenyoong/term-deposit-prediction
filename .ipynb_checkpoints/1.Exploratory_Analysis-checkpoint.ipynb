```python
# Assuming 'df' is your DataFrame
features = ['day', 'duration', 'campaign', 'previous', 'pdays']

# Custom color palette
custom_palette = sns.color_palette("Set2")

# Plot histograms for each feature
plt.figure(figsize=(20, 16))

for i, feature in enumerate(features, start=1):
    # Create subplots
    plt.subplot(3, 3, i)
    
    # Plot histogram with custom colors
    sns.histplot(df[feature], bins=20, kde=True, color=custom_palette[i-1])
    
    # Add title to the subplot
    plt.title(f'{feature.capitalize()} Distribution')

# Adjust layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```

**Day Distribution:** Shows which day of the month clients were contacted

**Duration Distribution:** Represents the contact duration with clients

**Campaign Distribution:** Indicates the number of contacts performed during this campaign for each client.

**Previous Distribution:** Reflects the number of contacts before the campaign with an each client.

**Pdays Distribution:** Shows the number of days passed by, after the client was last contacted from a previous campaign.

### 4. Correlation between target variable and independent variables

Encode categorical variables in the DataFrame using predefined mappings to prepare the data for visualizing the correlation between the target variable ('Target') and independent variables.

```python
# Define the mapping dictionary
mapping_dict = {
    "job": {"unknown": -1, "blue-collar": 1, "management": 2, "technician": 3, "admin.": 4, "services": 5,
            "retired": 6, "self-employed": 7, "entrepreneur": 8, "unemployed": 9, "housemaid": 10, "student": 11},
    "marital": {"single": 1, "married": 2, "divorced": 3},
    "education": {"unknown": -1, "primary": 1, "secondary": 2, "tertiary": 3},
    "default": {"no": 0, "yes": 1},
    "housing": {"no": 0, "yes": 1},
    "loan": {"no": 0, "yes": 1},
    "contact": {"unknown": -1, "cellular": 1, "telephone": 2},
    "month": {"jan": 1, "feb": 2, "mar": 3, "apr": 4, "may": 5, "jun": 6, "jul": 7, "aug": 8, "sep": 9, "oct": 10,
              "nov": 11, "dec": 12},
    "poutcome": {"unknown": -1, "failure": 0, "success": 1, "other": 2},
    "Target": {"no": 0, "yes": 1}
}

# Replace values using the mapping dictionary
df2 = df.replace(mapping_dict)

df2.head()
```

```python
# Selecting only numeric columns
numeric_columns = df2.select_dtypes(include='number')

plt.figure(figsize=(20, 15))
sns.heatmap(numeric_columns.corr(), linewidths=0.1, fmt='0.0%', linecolor='black', annot=True, cmap='magma')
plt.yticks(rotation=0)
plt.show()
```

- The heatmap above visualizes the correlation matrix of the dataset.
- The annotations on the heatmap provide the exact correlation percentages between the variables.
- This visualization can help identify which features are most related to each other.

## Features Engineering

By now, we should obtain some preliminary insights about how each important variable would affect the success rate, we will select the features that will be used in success prediction in the next part.

```python
df.head()
```

```python
columns_of_interest = ['age', 'job', 'marital', 'education', 'default', 'balance', 'housing', 'loan', 'contact', 'day', 'month', 'duration', 'campaign', 'pdays', 'previous', 'poutcome']

# Create a new DataFrame with selected columns
features = df[columns_of_interest]

# Display the first few rows of the new DataFrame
features.head()
```

### 1. Dropping the 'contact' column from the dataset.

Based on my understanding, the communication method ('Contact' column) is unlikely to play a significant role in model building. This is because there seems to be no meaningful relationship between the contact type and the outcome variable related to banking. As a result, I plan to drop the 'Contact' column from the model-building process.

```python
# Drop the 'contact' column from the DataFrame
features = features.drop(['contact'], axis=1)

# Display the first 5 rows of the updated DataFrame
print(features.head(5))
```

### 2. Replacing 'unknown' values with 'others' in the 'poutcome' column.

```python
features['poutcome'].value_counts()
```

In the 'poutcome' column, the values 'other' and 'unknown' don't provide significant insights. I will replace 'unknown' with 'others'.

```python
# Replace encoded value -1 with 2 in 'poutcome' column
df['poutcome'].replace("unknown", "others", inplace=True)

# Display the first 4 rows of the 'poutcome' column
df['poutcome'].head(5)
```

### 3. Generating Dummy Variables

```python
object_columns = ["job", "marital", "education", "default", "housing", "loan", "contact", "month", "poutcome"]

features_one_hot = pd.get_dummies(features,columns=object_columns)
features_one_hot
```

Use the function `get_dummies` and `features`dataframe to apply OneHotEncoder to the columns. Assign the value to the variable features_one_hot, display the results using the method head. Your result dataframe must include all features including the encoded ones.

```python
features_one_hot.columns
```

Now that our `features_one_hot` dataframe only contains numbers cast the entire dataframe to variable type `float64`

```python
features_one_hot = features_one_hot.astype('float64')
```

We can now export it to a **CSV** for the next part.

```python
features_one_hot.to_csv('bank-full2.csv', index=False)
```
